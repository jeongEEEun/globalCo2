```python
import pandas as pd
import numpy as np
```

```python
co2_weekly = pd.read_csv('co2_weekly_mlo.csv')
co2_weekly
```

```python
## [필요한 컬럼만 추출]
co2_weekly = co2_weekly[['year', 'month', 'day', 'average', 'increase since 1800']]
```

```python
## [컬럼명 변경] 영어 -> 한글
co2_weekly.columns = ['기준연도', '기준월', '기준일', '평균값', '1800년_이후_증가값']
```

```python
## [결측치 재설정] -999.99 -> NaN
co2_weekly.loc[co2_weekly.평균값 == -999.99, '평균값'] = np.nan
```

```python
co2_weekly.to_csv('주간_이산화탄소_농도.csv', encoding='utf-8', index=False)
```
